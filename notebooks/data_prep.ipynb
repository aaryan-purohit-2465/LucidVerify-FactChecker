```python
# FULL preprocessing + splits — run this single cell
import pandas as pd
import re
from sklearn.model_selection import train_test_split

# 1) load
fake = pd.read_csv('../data/Fake.csv')
true = pd.read_csv('../data/True.csv')

# 2) label
fake['label'] = 0
true['label'] = 1

# 3) concat
df = pd.concat([fake, true], ignore_index=True)

# 4) basic cleanup: drop dups, drop na, keep only required columns
df.drop_duplicates(inplace=True)
df.dropna(subset=['text'], inplace=True)        # be explicit: require 'text' column
# if dataset uses 'title' or 'text' differently, adapt next line:
if 'text' not in df.columns and 'title' in df.columns:
    df = df.rename(columns={'title':'text'})
df = df[['text','label']]

# 5) light text cleaning (fast)
cleaner = re.compile(r"http\S+|\s+")
df['text'] = df['text'].apply(lambda t: cleaner.sub(" ", str(t)).strip())

# 6) shuffle
df = df.sample(frac=1, random_state=42).reset_index(drop=True)

# 7) splits (80/10/10) with stratify on label
train_df, temp_df = train_test_split(df, test_size=0.2, random_state=42, stratify=df['label'])
val_df, test_df = train_test_split(temp_df, test_size=0.5, random_state=42, stratify=temp_df['label'])

# 8) save splits
train_df.to_csv('../data/train_clean.csv', index=False)
val_df.to_csv('../data/val_clean.csv', index=False)
test_df.to_csv('../data/test_clean.csv', index=False)

# 9) show sizes
print("Sizes (train, val, test):", len(train_df), len(val_df), len(test_df))
```
